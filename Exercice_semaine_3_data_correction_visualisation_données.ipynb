## Partie1 : Visualisation d’une fonction

```python
# 1.Définir dans python la fonction 
# f (a) = exp(-a) x cos(2π x a)  [x est le signe de la multiplication]
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(a):
    y = np.exp(-a) * np.cos(2*np.pi * a) 
    return y
```

```python
f(2)
```

```python
a1 = np.arange(0.0, 5.0,0.05)
len(a1)
```

```python
a2 = np.linspace(0, 5, 50)
a2
```

```python
# 2.Tracer la fonction y = f(a) dans l’intervalle [0, 5]. 
plt.figure() 
plt.plot(a1,f(a1))
plt.xlabel("a")
plt.ylabel("f(a)")
```

```python
# 3.L’équation f(a) = 0 a-t-elle des solutions entières dans l’intervalle [0,100] ?  
for i in np.arange(0,101,1):
    y = f(i)
    if y==0:
        print(i)
    else:
        print("no solution")
# Réponses:pas de solution entière
```
